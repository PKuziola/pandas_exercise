### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
round(users[['occupation','age']].groupby('occupation').mean(),2).sort_values('age',ascending=False)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
for index, row in users.iterrows(): 
    if row['gender']=='M':        
        users.loc[index,'male_int'] = 1        
    else:        
        users.loc[index,'male_int'] = 0        

users_male_ratio = users[['occupation','male_int']].groupby('occupation').agg(['count','sum'])
users_male_ratio['male_ratio']= round(users_male_ratio[('male_int', 'sum')]/users_male_ratio[('male_int', 'count')],4)*100
users_male_ratio.sort_values('male_ratio',ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users[['occupation','age']].groupby('occupation').aggregate(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users
round(users[['occupation','age','gender']].groupby(['occupation','gender']).mean(),2)
```

### Step 8.  For each occupation present the percentage of women and men

```python
users_percentage_gender = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

users_percentage_count = users.groupby(['occupation']).agg('count')

final_dataframe = round(users_percentage_gender.div(users_percentage_count, level = "occupation") * 100,2)

final_dataframe['gender']

 
```
